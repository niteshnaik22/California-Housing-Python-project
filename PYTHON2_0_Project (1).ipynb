10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
#Two list of total bedrooms and bed size is prepared, to form a new column in the dataset.
total_bedrooms = [(housing['total_bedrooms'] <=10),(housing['total_bedrooms'] >=11)&(housing['total_bedrooms'] <=1000),(housing['total_bedrooms'] > 1000)]
bed_size= ['small','medium','large']
#new data frame named 'total_bedroom_size' is formed using select function.
housing['total_bedroom_size']=np.select(total_bedrooms,bed_size)
housing
```

When we compare the new column total bedroom size with the total bedroom column we see that 1000 above are represented as large, below 1000 and above 11 is represenetd as medium and below 11 as small.
